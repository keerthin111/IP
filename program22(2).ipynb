```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt

# Load the image
img = io.imread("img3.jpg")

# Prepare the figure
fig, axs = plt.subplots(1, 3, figsize=(16,4))

# Display the image
axs[0].imshow(img, cmap="gray")

# Plot the first histogram
axs[1].hist(
    img.ravel(),            # The image must be flattened to use the function hist
    bins=range(0,256,2)     # Define 128 bins between 0 and 255
)
axs[1].set_xlabel("Intensities")
axs[1].set_ylabel("Number")

# Plot the second histogram
axs[2].hist(
    img.ravel(),            # The image must be flattened to use the function hist
    bins=range(0,256,16)    # Define 16 bins between 0 and 255
)
axs[2].set_xlabel("Intensities")
axs[2].set_ylabel("Number")

# Show the figure
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('img1.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(histogram, color='k')
plt.show()
```

```python
import cv2  # Open CV
from matplotlib import pyplot as plt  #Matplotlib for plotting  
  
# Read the input image
img = cv2.imread('img5.jpg',0) 
plt.imshow(img)
plt.show()
  
# Calculate the frequency of pixels in the brightness range 0 - 255
histr = cv2.calcHist([img],[0],None,[256],[0,256]) 
  
# Plot the histogram and display
plt.plot(histr) 
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
img = np.array(Image.open('img5.jpg'))
img_flat = img.flatten()
plt.hist(img_flat, bins=200, range=[0, 256])
plt.title("Number of pixels in each intensity value")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.show()
```
