```python
import cv2
import matplotlib.image as mping
import matplotlib.pyplot as plt
#reading image files
img1=cv2.imread('ff.jpg')
img2=cv2.imread('fish1.jpg')
#applying numpy adddition on image
fimg1=img1+img2
plt.imshow(fimg1)
plt.show()
#saving the output image
cv2.imwrite('output.jpg',fimg1)
fimg2=img1-img2
plt.imshow(fimg2)
plt.show()

cv2.imwrite('output.jpg',fimg2)
fimg3=img1*img2
plt.imshow(fimg3)
plt.show()

cv2.imwrite('output.jpg',fimg3)
fimg4=img1/img2
plt.imshow(fimg4)
plt.show()

cv2.imwrite('output.jpg',fimg4)
```
